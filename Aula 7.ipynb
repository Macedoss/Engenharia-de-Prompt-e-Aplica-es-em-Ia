```python
 =========================================
# CALCULADORA SIMPLES EM PYTHON
# =========================================

# A função "somar" recebe dois números
# e retorna a soma deles
def somar(a, b):
    return a + b


# A função "subtrair" recebe dois números
# e retorna a subtração
def subtrair(a, b):
    return a - b


# A função "multiplicar" recebe dois números
# e retorna a multiplicação
def multiplicar(a, b):
    return a * b


# A função "dividir" recebe dois números
# e retorna a divisão
def dividir(a, b):
    # Aqui fazemos uma verificação para evitar
    # erro ao dividir por zero
    if b == 0:
        return "Erro: não é possível dividir por zero."
    return a / b


# Exibe o título da calculadora
print("=== CALCULADORA SIMPLES ===")

# Mostra as opções disponíveis para o usuário
print("Escolha uma operação:")
print("1 - Somar")
print("2 - Subtrair")
print("3 - Multiplicar")
print("4 - Dividir")

# input() serve para receber dados digitados pelo usuário
operacao = input("Digite o número da operação desejada: ")

# float() converte o valor digitado para número com casas decimais
# Exemplo: 10.5
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Agora usamos if / elif / else para descobrir
# qual operação o usuário escolheu
if operacao == "1":
    resultado = somar(num1, num2)
    print("Resultado:", resultado)

elif operacao == "2":
    resultado = subtrair(num1, num2)
    print("Resultado:", resultado)

elif operacao == "3":
    resultado = multiplicar(num1, num2)
    print("Resultado:", resultado)

elif operacao == "4":
    resultado = dividir(num1, num2)
    print("Resultado:", resultado)

else:
    # Caso o usuário digite algo inválido
    print("Opção inválida.")
```
