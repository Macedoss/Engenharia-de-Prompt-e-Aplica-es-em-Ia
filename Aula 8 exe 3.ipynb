```python
import time, random, smtplib
from email.mime.text import MIMEText

# Configurações de limite e acesso
C = {"lim": 70, "int": 5, "rem": "seu@email.com", "des": "para@email.com", "pass": "senha"}

def enviar_email(v):
    try:
        msg = MIMEText(f"Alerta: {v}!")
        msg['Subject'], msg['From'], msg['To'] = "🚨 ALERTA", C['rem'], C['des']
        # Conexão segura com servidor SMTP
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as s:
            s.login(C['rem'], C['pass'])
            s.send_message(msg)
        print("✅ E-mail enviado.")
    except Exception as e: print(f"❌ Erro: {e}")

try:
    while True:
        v = random.randint(20, 100) # Gera valor aleatório
        status = "🚨 ALERTA!" if v > C['lim'] else "Normal"
        print(f"Leitura: {v} - {status}")

        # Dispara e-mail se passar do limite
        if v > C['lim']: enviar_email(v)

        time.sleep(C['int']) # Pausa entre leituras
except KeyboardInterrupt: print("Parado.")
```
